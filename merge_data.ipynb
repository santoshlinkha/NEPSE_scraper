```python
import os
import glob
import pandas as pd

path = './'
os.chdir(path)
all_csv_files = glob.glob('*.csv')

database = []

for file in all_csv_files:
    df = pd.read_csv(path + file)[::-1]
    database.append(df)
```

```python
data = pd.concat(database, ignore_index=True)
data = data.iloc[:, :-1]
data.to_csv("database.csv")
```
